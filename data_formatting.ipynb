```python
import pandas as pd
```

```python
heart = pd.read_csv("archive/heart.csv")
```

```python
normalized_heart=(heart-heart.min())/(heart.max()-heart.min())
```

```python
heart
```

```python
normalized_heart = normalized_heart.drop(labels = 'output', axis = 1)
```

```python
normalized_heart.head()
```

```python
X = normalized_heart.values.flatten().tolist()
Y = heart['output'].values.tolist()
# Y needs to have values in [1,...] so we will change 0's to 2's
for i in range(len(Y)) :
    if Y[i] == 0 :
        Y[i] = 2
```

```python
number_of_features = int(len(X)/len(Y))
```

```python
number_of_features
```

```python
with open('heart.txt', 'w') as f:
    f.write("X = [")
    for x in range(len(X)) :
        f.write("%s" % X[x])
        if ((x+1)%number_of_features==0) & (x < len(X)-1) :
            f.write("; ")
        elif (x < len(X)-1) :
            f.write(" ")
    f.write("]\n")
    
    f.write("Y = [")
    for y in range(len(Y)) :
        f.write("%s" % Y[y])
        if y< (len(Y) - 1) :
            f.write(", ")
    f.write("]")
```
